```python
labdir = "/facebook_clean_data/"
import csv

with open("tvshow_edges.csv") as csvfile:
    facebook_clean_data = list(csv.reader(csvfile)) 
```

```python
def processing_data():
  dic = {}
  for i in facebook_clean_data[1:]:
    if not i[0] in dic:
      dic[i[0]] = [i[1]]
    else:
      dic[i[0]].append(i[1])
  data_set = list(dic.values())
  return data_set
data_set = processing_data()
```

```python
def createC1(dataSet):
    C1=[] # Put the frequent candidate 1-itemset
    for transaction in dataSet:
        for item in transaction:
            if not {item} in C1:
                C1.append({item})
    C1.sort()
    return list(map(frozenset,C1))
```

```python
C1=createC1(data_set)
```

```python
def scanD(D,Ck,minSupport):
    ssCnt={}
    for tid in D:
        for can in Ck:
            if can.issubset(tid): #can in tid or not
                ssCnt[can]=ssCnt.get(can,0)+1
    numItems=float(len(D))
    reList=[]
    supportData={}
    for key in ssCnt:
        support=ssCnt[key]/numItems
        supportData[key]=support
        if support >= minSupport:
            reList.append(key)
    return reList,supportData
             
```

```python
#频繁k项集生成候选k+1项集
def aprioriGen(Lk):
    Ck=[] #put Candidate k+1 itemsets on Ck
    lenLk=len(Lk) #频繁k项集的长度
    for i in range(lenLk):
        for j in range(i+1,lenLk):
            L1=Lk[i]
            L2=Lk[j]
            C=Lk[i]|Lk[j]
            if not C in Ck and (len(C)==len(Lk[0])+1):
                Ck.append(Lk[i]|Lk[j])
    return Ck
```

```python
def apriori(D,minSupport =0.02):
    C1=createC1(D) #generate candidate one itemset
    L1,supportData=scanD(D,C1,minSupport) #generate frequent one itemset and support table
    L=[L1]  #put L1 in L
    k=2  #number of itemset,let initial value of k be 2
    while (len(L[-1])>0): #till the last item of L be null
        Ck=aprioriGen(L[-1]) #generate candidate k itemset
        Lk,supK=scanD(D,Ck,minSupport) #generate frequent k itemset
        supportData.update(supK) #update supportData
        L.append(Lk) #update L
        k=k+1
    return L,supportData
```

```python
L,supportData=apriori(data_set,minSupport=0.02) 
```

```python
def generate_big_rules(L, support_data, min_conf):
    big_rule_list = []
    sub_set_list = []
    for i in range(0, len(L)):
        for freq_set in L[i]:
            for sub_set in sub_set_list:
                if sub_set.issubset(freq_set):
                    conf = support_data[freq_set] / support_data[freq_set - sub_set]
                    big_rule = (freq_set - sub_set, sub_set, conf)
                    if conf >= min_conf and big_rule not in big_rule_list:
                        # print freq_set-sub_set, " => ", sub_set, "conf: ", conf
                        big_rule_list.append(big_rule)
            sub_set_list.append(freq_set)
    return big_rule_list
# ToDo: Write you code here.
```
